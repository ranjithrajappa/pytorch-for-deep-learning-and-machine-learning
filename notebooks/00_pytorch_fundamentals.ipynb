#Tensor Datatypes
### Note: Tensor datatypes is one of the 3 big errors you'll run into with PyTorch and deep learning:

1. Tensors not right datatype
2. Tensors not right shape
3. Tensors not right device

```python
# Float 32 tensor (default)
float_32_tensor = torch.tensor([1.0,2.0,3.2],
                               dtype=None,# what datatype is the tensor ex:float32(default) or float16
                               device=None,# what device is your tensor on ex: cpu(default) or cuda
                               requires_grad=False) # whether or not to track gradients with this tensor's operations
print(float_32_tensor)
print(float_32_tensor.dtype)

# The default datatype is going to come out as float32 even if dtype is specified as None ( and float type numbers are used to create tensor).
# 'FP32' or 'float32' or '32-bit floating point' or 'single-precision floating point format'
# 'FP16' or 'float16' or '16-bit floating point' or 'half-precision floating point format'
# 'FP64' or 'float64' or '64-bit floating point' or 'double-precision floating point format'

#In computer science, precision refers to how accurately a computer can represent and calculate numbers, especially decimal or fractional values.
# It’s about how many digits (or bits) are used to store a number, which determines how close the stored value is to the actual mathematical value.
```

```python
# Convertin float32 to another dtype
float_16_tensor_a=float_32_tensor.type(torch.float16)
float_16_tensor_b=float_32_tensor.type(torch.half)
print(float_16_tensor_a,float_16_tensor_a.dtype)
print(float_16_tensor_b,float_16_tensor_b.dtype)
```

```python
#defining the dtype while creating the tensor, if it shouldn't be default float32
float_64_tensor= torch.tensor([2.0,3.0,4.0],
                                dtype=torch.float64,
                                )
float_64_tensor.dtype
```

```python
d=float_32_tensor*float_64_tensor
e=float_32_tensor+float_64_tensor
print(float_32_tensor,float_64_tensor)
print(d,d.dtype)
print(e,e.dtype)
```


Getting information from tensors(Tensor attributes):

1.	Tensors not right datatype – to get datatype from a tensor, type tensor.dtype

2.	Tensors not right shape – to get shape from a tensor, type tensor.shape

3.	Tensors not right device – to get device from a tensor, type tensor.device
